```python
import pandas as pd
import numpy as np
import tensorflow as tf

from imblearn.under_sampling import RandomUnderSampler
import random
random.seed(42)
```

```python
data=pd.read_csv("akakce_paired"+".csv",encoding = "ISO-8859-1")
data.head()
```

```python
websites=["akakce","amazon","arabam","donanimhaber","haberturk","mgm","nefisyemektarifleri","pazarama","trendyol"]

datas=[]
for website in websites:
    data=pd.read_csv(website+"_paired.csv")
    # Perform random under sampling
    rus = RandomUnderSampler(random_state=42)
    data_rus, _ = rus.fit_resample(data, data["identical"])
    data_rus.reset_index(inplace=True)
    print(website,data_rus.shape)
    datas.append(data_rus)
data=pd.concat(datas,ignore_index=True).drop("index",axis=1)
print(data.shape)
data.head()
```

```python

# Load the model

model = tf.keras.applications.ResNet50V2(input_shape=(256,256,3),
                                               include_top=False,
                                               weights='imagenet')

model=tf.keras.Model(model.input,tf.keras.layers.GlobalMaxPooling2D()(model.output))

def preprocess_and_predict(image_path,model):
    image = tf.keras.utils.load_img(image_path,target_size=(256,256))
    image_array = np.array(image)
    image_array = np.expand_dims(image_array, axis=0)
    predictions = model.predict(tf.keras.applications.resnet_v2.preprocess_input(image_array),verbose=0).flatten()
    return predictions

# Apply the function to the images

url1_features = data.apply(lambda x: preprocess_and_predict("data/"+x["url1_dir"]+"/ss.png", model), axis=1)
url1_features=np.array(url1_features.tolist())


url2_features = data.apply(lambda x: preprocess_and_predict("data/"+x["url2_dir"]+"/ss.png", model), axis=1)
url2_features=np.array(url2_features.tolist())
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(np.concatenate([url1_features,url2_features],axis=1),data["identical"],test_size=0.2)
```
