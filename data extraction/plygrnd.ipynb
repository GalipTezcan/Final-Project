```python
from input_crawler import InputCrawler
from data_extractor import Labeler
from web_crawler import WebCrawler
import csv
```

```python
base_url = "https://www.pazarama.com/"  # Replace with your target URL
csvf="pazarama"
max_depth = 2  # Set your desired crawl depth

crawler = WebCrawler(base_url, max_depth,50)
crawler.crawl(base_url)

urllist=crawler.visited
```

```python
len(list(urllist))
```

```python
Labeler(list(urllist),csvf+".csv")
```

```python
import csv
urllist=[]
with open(csvf+".csv", mode='r') as file:
    # Create a CSV reader object
    csv_reader = csv.reader(file)

    # Read the header row
    headers = next(csv_reader)
    print("Headers:", headers)
    

    # Read and print the rest of the rows
    for row in csv_reader:
        if len(row)>0:
            urllist.append(row[0])
        print("Row:", row)
```

```python
len(urllist)
```

```python
urllist=list(urllist)
with open(csvf+"_dirs.csv", 'w') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["url","dir"])
    for i in range(len(urllist)):
        InputCrawler(urllist[i],i)
        writer.writerow([urllist[i],i])
```
